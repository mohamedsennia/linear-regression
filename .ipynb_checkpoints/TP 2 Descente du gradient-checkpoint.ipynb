# Descente du Gradient : Préparation des fonctions

1- Calcul du coût

Cette fonction servira à calculer le cout $J(\theta_0,\theta_1)$

Elle prendra l'ensemble de données d'apprentissage en entrée ainsi que les paramètres définis initialement

```python
def computeCostNonVect(X, y, theta):
    # idéalement, tracer le coût à chaque itération pour s'assurer que la descente du gradient est correcte
    j=0
    for i in range(len(X)):
        h=theta[0][0]+theta[1][0]*X[i]
        h=h-y[i]
        h=pow(h,2)
        j=j+h
    m=len(X)*2
    j=j/m
    return j
    # calculer le coût avec et sans vectorisation, 
    # comparer le temps de traitement
    
```

```python
def computeCost(X, y, theta):
    # idéalement, tracer le coût à chaque itération pour s'assurer que la descente du gradient est correcte
    j=0
    for i in range(len(X)):
        h=np.dot(X[i,:],theta)
        h=h-y[i]
        h=pow(h,2)
        j=j+h
    m=len(X)*2
    j=j/m
    return j
    # calculer le coût avec et sans vectorisation, 
    # comparer le temps de traitement
    
  
```

2- Fonction de la descente du gradient

Cette fonction mettra à jour les paramètres $\theta_0,\theta_1$ jusqu'à convergence: atteinte du nombre d'itérations max, ou dérivée assez petite.

```python
def gradientDescent(X, y, theta, alpha, iterations):
    # garder aussi le cout à chaque itération 
    # pour afficher le coût en fonction de theta0 et theta1
    cost=np.zeros(iterations)
    thetas0=np.zeros(iterations)
    thetas1=np.zeros(iterations)
    m = len(y) 
    for i in range(iterations):
        thetas0[i]=theta[0]
        thetas1[i]=theta[1]
        cost[i]=computeCost(X,y,theta)
        h = np.transpose(np.matmul(X,theta))
        theta[0] = theta[0] - (alpha * ((1/m) * np.sum(h - y)))
        theta[1] = theta[1] - (alpha * ((1/m) * np.sum((h - y) * X[:,1])))
    return theta,cost,thetas0,thetas1
```

# Descente du Gradient : Appel des fonctions

Initialisation de $\theta_0$ et $\theta_1$

```python
theta = np.zeros((2, 1))
```

Calculer le cout initial

```python
initialCost=computeCost(X, y, theta)
print(initialCost)
```

Appel des la fonction de calcul du gradient

```python
# paramètres
iterations = 1500;
alpha = 0.01;
# Appel
cost=np.zeros(iterations)
thetas0=np.zeros(iterations)
thetas1=np.zeros(iterations)
theta,cost,thetas,thetas1 = gradientDescent(X, y, theta, alpha, iterations);
```

# Régression linéaire à plusieurs variables

Proposer, l'aide des fonctions définies precedemment, une regression linéaire lorsque le nombre de descripeurs est supérieur à 1, en utilisant la base d'apprentissage suivante

```python
# données
dataMulti = np.genfromtxt('dataMulti.csv', delimiter=',', dtype=int)
dataMulti.shape
```

Pour cette ensemble de données, nous voudrions prédire le prix d'une maison (3ème colonne de dataMulti) à partir de :
- sa supérficie (1ère colonne)
- son nombre de chambres (2ème colonne)

```python
# d'abord créer X et y
intercept=np.ones((dataMulti.shape[0],1))
XM=np.column_stack((intercept,dataMulti[:,0:2]))
yM = dataMulti[:, 2]
theta=np.zeros((3,1))
```

```python
# redéfinissez vos fonctions de coût si cela est necessaire
#pas necessaire dans ce cas
```

Mise à l'échelle des données

```python

x1 = XM[:,1]
x2 = XM[:,2]

maxX1=x1.max()
maxX2=x2.max()
minX1=x1.min()
minX2=x2.min()
sX=np.copy(XM)
sY=np.copy(yM)
for i in range(len(XM)):
    sX[i][1]=(sX[i][1]-minX1)/(maxX1-minX1) 
    sX[i][2]=(sX[i][2]-minX2)/(maxX2-minX2)
maxY=yM.max()
minY=yM.min()
for i in range(len(sY)):
    sY[i]=(sY[i]-minY)/(maxY-minY)
```

Normalisation des données en moyenne = 0

```python
sx1= np.array(sX[:,1])
sx2= np.array(sX[:,2])
sx=  np.zeros((sx1.shape[0], 2))

for i in range (sx.shape[0]):
    sx[i][0]=sx1[i]
    sx[i][1]=sx2[i]
for i in range(sX.shape[0]):
    sX[i][0]=sX[i][0]/np.std(sx)
    sX[i][1]=sX[i][1]/np.std(sx)
for i in range(len(yM)):
    sY[i]=sY[i]/np.std(yM)
```

Appliquer la descente du gradient à plusieurs variables

```python
def gradientDescent2(X, y, theta, alpha, iterations):
    m = len(y)
    for i in range(iterations):
        h = np.dot(X,theta)
        theta[0] = theta[0] - (alpha * ((1/m) * np.sum(h - y)))
        theta[1] = theta[1] - (alpha * ((1/m) * np.sum((h-y)* X[:, 1])))
        theta[2] = theta[2] - (alpha * ((1/m) * np.sum((h - y) * X[:,2])))
    return theta
```

Comparer la descente du gradient avec et sans normalisation
- **Exemple**: voir la difference dans le temps de calcul

```python
s= time.time()
gradientDescent2(XM,yM,theta,alpha,iterations)
t=time.time()
print("temps de calcul sans normalisation")
print(t-s)
s= time.time()
gradientDescent2(sX,sY,theta,alpha,iterations)
t=time.time()
print("temps de calcul avec normalisation")
print(t-s)
```
